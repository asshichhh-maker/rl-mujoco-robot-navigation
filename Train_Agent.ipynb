```python
import gymnasium as gym
from stable_baselines3 import SAC

env = gym.make("Walker2d-v5")

model = SAC(
    "MlpPolicy",
    env,
    verbose=1,
    learning_rate=3e-4,
    buffer_size=1000000,
    batch_size=256,
    gamma=0.99,
)

model.learn(total_timesteps=1000000)

model.save("walker2d_simple_model")

print("Training complete ")
```

Vedio Record



```python
import gymnasium as gym
from stable_baselines3 import SAC
import imageio

# Timesteps to compare
checkpoints = [0, 10000, 50000, 100000, 500000, 1000000]

for checkpoint in checkpoints:

    print(f"\nLoading model at {checkpoint} timesteps...")

    model = SAC.load(f"walker2d_{checkpoint}")

    env = gym.make("Walker2d-v5", render_mode="rgb_array")

    frames = []

    obs, info = env.reset()

    print(f"Recording video for {checkpoint} timesteps...")

    for i in range(1500):

        action, _ = model.predict(obs)

        obs, reward, terminated, truncated, info = env.step(action)

        frame = env.render()
        frames.append(frame)

        if i % 100 == 0:
            print(f"Step: {i}")

        if terminated or truncated:
            obs, info = env.reset()

    env.close()

    video_name = f"walker2d_{checkpoint}.mp4"

    print(f"Saving {video_name}...")

    imageio.mimsave(video_name, frames, fps=30)

    print(f"{video_name} saved successfully!")

print("\nAll videos generated!")
```
